```bash
#start/restart orientdb

/i dup orientdb
```

```bash
#install orientdb console
if [ ! -e /orientdb-console/bin/console.sh ]; then
    mkdir -p /orientdb-console
    cd /orientdb-console
    mkdir -p bin
    /i cp orientdb:/orientdb/bin/console.sh ./bin/
    mkdir -p lib
    /i cp orientdb:/orientdb/lib/ .
    #mkdir -p config/cert
    #/i cp orientdb:/orientdb/config/cert/orientdb-console.ks ./config/cert/
fi

# save dbaccess config (connection command)
cat >$HOME/.orientdb/connString <<EOF
CONNECT remote:orientdb/demodb admin admin;
EOF

#create wrapper script for use from jupyter
if [ ! -e /orientdb-console/bin/consoleWrapper.sh ]; then
cat >/orientdb-console/bin/consoleWrapper.sh <<EOF
    #!/bin/bash
    uuid=\$(uuidgen)
    sfile=/tmp/oc-\$uuid
    sfiler=/tmp/oc-\$uuid-resp
    cat $HOME/.orientdb/connString > \$sfile 
    if [ "x\$1" = "xf" ]; then
        cat \$f | tr -d '\n' >> \$sfile
    elif [ "x\$1" = "xd" ]; then 
        echo "\$2" | tr -d '\n' >> \$sfile
        cat  \$sfile
        printf "\n ========= debug ========= \n"
    else 
        echo "\$1" | tr -d '\n' >> \$sfile
    fi    
    /orientdb-console/bin/console.sh \$sfile > \$sfiler
    cat \$sfiler 
EOF
chmod a+x /orientdb-console/bin/consoleWrapper.sh
fi

#install uuidgen
if [ ! -e /run/uuidd ]; then 
    mkdir -p /run/uuidd
    apt-get install -y  uuid-runtime 
fi
ln -sf /orientdb-console/bin/consoleWrapper.sh /oc
```
